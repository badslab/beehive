```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
n = 500
d = pd.concat([
    pd.DataFrame(dict(
        x = np.random.normal(10, 2, (n, )), y = np.random.normal(9, 2, (n, )), z = 1)),
    pd.DataFrame(dict(
        x = np.random.normal(8, 2, (n, )), y = np.random.normal(10, 4, (n, )), z = 2)),])

d.head()
```

```python
sns.scatterplot(x='x', y='y', hue='z', data=d, palette='Set1', alpha=0.5);
```

```python
sns.h(data=d, x='x', y='y')
```
